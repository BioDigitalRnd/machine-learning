```python
# import necessary modules and algorithms
import pandas as pd
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

```python
# load all patient files (1-53) and store in dictionary
patients = {}
for m in range(1, 54):
    patients[f"patient{m}"] = pd.read_csv(
        f'bidmc_{str(m).zfill(2)}_Numerics.csv',
        na_values='NaN'
    )
```

```python
x = 1
len(patients[f'patient{x}']) # returns the amount of rows in the profile
```

```python
# add an id column for each patient (1-53)
for id in range(1, 54):
    patients[f'patient{id}']['id'] = [id]*len(patients[f'patient{id}'])
```

```python
# Extract age and gender data from text files

patient_age = {}
patient_gender_M = {}
patient_gender_F = {}

for x in range(1, 54):
    with open (f'bidmc_{str(x).zfill(2)}_Fix.txt') as pt:
        for rec in pt:
            if rec[0:3] == 'Age':
                if not rec[5:7] == 'Na':
                    patient_age[f'patient{x}'] = rec[5:7]
                else:
                    patient_age[f'patient{x}'] = 'NaN'
            if rec[0:3] == 'Gen':
                if rec[8] == 'M':
                    patient_gender_M[f'patient{x}'] = 1
                    patient_gender_F[f'patient{x}'] = 0
                elif rec[8] == 'F':
                    patient_gender_F[f'patient{x}'] = 1
                    patient_gender_M[f'patient{x}'] = 0
                else:
                    print(f'patient{x}')
                    patient_gender_F[f'patient{x}'] = 'NaN'
                    patient_gender_M[f'patient{x}'] = 'NaN'
                

# create new columns for age and gender and assign them to each patient
for dat in range(1,54):
    patients[f'patient{dat}']['age'] = patient_age[f'patient{dat}']
    patients[f'patient{dat}']['gender_M'] = patient_gender_M[f'patient{dat}']
    patients[f'patient{dat}']['gender_F'] = patient_gender_F[f'patient{dat}']
```

```python
#identify which patients have missing information
import numpy as np
patient_features = ['time', 'hr', 'pulse', 'resp', 'spo2', 'age', 'gender_M', 'gender_F']
indicator = 1
for patient in patients:
    patients[patient].replace('NaN', np.nan, inplace=True)
    
    for feature in patient_features:
        if patients[patient][feature].isna().sum() > 0:
            print(f"patient{indicator}")
            print(patients[patient].isna().sum())
            break
    indicator = indicator + 1
```

```python
# remove rows where there are missing pulse values
for patient in patients:
    patients[patient].dropna(how='any', subset=['pulse'], inplace=True)
```

```python
# make new list of patient profiles concatinated together
f = 1
new_list= []
# patients[f'patient{f}']
for patient in patients:
    new_list.append(patients[patient])
```

```python
# Combine all patient profiles into one
result = pd.concat(new_list)
```

```python
# drop all the rows with missing resp values
result.dropna(how='any', subset=['resp'], inplace=True)
```

```python
# drop all rows with missing hr values
result.dropna(how='any', subset=['hr'], inplace=True)
```

```python
# Assign independent and dependent variables
X_01 = result.drop(columns=['resp'])
y_01 = result['resp']
```

```python
X_01.shape
```

Apply algorithms to datasets and score using cross validation method

```python


ranfor = RandomForestRegressor(random_state=1)
#ranfor.fit(X_01, y_01)

from sklearn.model_selection import cross_val_score, StratifiedKFold
cross_val_score(ranfor, X_01, y_01, cv=StratifiedKFold(shuffle=True,
                                                       n_splits=4,
                                                       random_state=1)).mean()
```

```python
linreg = LinearRegression()

cross_val_score(linreg, X_01, y_01, cv=StratifiedKFold(shuffle=True,
                                                      n_splits=8,
                                                      random_state=1)).mean()
```

```python
gradbst = GradientBoostingRegressor()

from sklearn.model_selection import cross_val_score, StratifiedKFold
cross_val_score(gradbst, X_01, y_01, cv=StratifiedKFold(shuffle=True,
                                              n_splits=8,
                                              random_state=1)).mean()
```

```python
from sklearn.svm import SVR

svr = SVR()

cross_val_score(svr, X_01, y_01, cv=StratifiedKFold(shuffle=True,
                                              n_splits=8,
                                              random_state=1)).mean()
```

Some optimisation on the above algorithms could be done (GridSearchCV/RandomizedSearchCV) to improve performance, but first the data would need to be split into train and test sets.
