## Note:
Split data before any preprocessing (including encoding and especially before imputation) to avoid data leakages.
The train set should be treated as if it is the only data you have to train the model, and the test set should be seen as data you will see in the future (at present you don't have access to).
The train and test sets should be completely independent of each other.
(Imputation before splitting will use the test data to fill missing values for the train data.)
This means the test set will have to go through its own preprocessing before being used for predictions/accuracy measurements.

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['binaryClass'])
y = df['binaryClass']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

It is good practise to split your data evenly (this data splits in 4 parts without remainders)
